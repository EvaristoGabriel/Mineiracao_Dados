```python
sns.boxplot(data=df, x='consome_comida_calorica', y='peso')
```

Podemos perceber que há uma maior dispersão dos valores de peso de pessoas que consomem comida calórica, tendo uma mediana de mais ou menos 85Kg.

Já no grupo de pessoas que não consomem comida calórica, temos uma mediana mais baixa, de aproximadamente 65Kg, e podemos perceber uma menor dispersão dos dados.

Podemos dizer então que pessoas que não consomem comida calórica tendem a possuir pesos mais baixos e com menos variabilidade.

```python
sns.boxplot(data=df, x='consome_comida_calorica', y='consumo_diario_agua')
```

Podemos perceber uma maior tendencia de pessoas que consomem comida calórica a consumir mais água, e pessoas que não consomem comida calórica não beber tanta água.

```python
sns.scatterplot(data=df, x='frequencia_atividade_fisica', y='peso')
```

Podemos observar que pessoas que praticam mais atividade física tendem a ter um peso menor do que as que não praticam

```python
sns.regplot(data=df, x='tempo_usando_eletronicos', y='frequencia_atividade_fisica', scatter_kws={'alpha': 0.5})
```

Podemos perceber que não há uma mudança muito grande na linha da regressão, isso mostra que esses dados não são associativos, isso é, o tempo que uma pessoa utiliza eletrônicos, não interfere na quantidade de atividade física que essa pessoa realiza.

Podemos constatar esse fato no gráfico abaixo, onde mostra que pessoas mais jovens tendem a utilizar mais o celular

```python
sns.regplot(data=df, x='tempo_usando_eletronicos', y='idade', scatter_kws={'alpha': 0.5})
```
